## Sayama Exercise 15.2
Calculate the following:
- The number of edges in a complete graph made of n nodes

"a complete graph is a simple undirected graph in which every pair of distinct vertices is connected by a unique edge." - Wikipedia. If we exclude the node n which can't be connected to itself, that very node has to be linked to n-1 nodes. Thus, the number of edges in a complete graph is: $$\frac{n \times \left(n-1\right)}{2}$$


- The number of edges in a regular graph made of n nodes each of which has degree k

"a regular graph is a graph where each vertex has the same number of neighbors". Handshaking Theorem states that the sum of degree of all the vertices is $2\times Edges$ in the graph, therefore:
$$E=\frac{n\times k}{2}$$

## Sayama Exercise 15.3
Determine whether the following graphs are planar or not:
- A complete graph made of four nodes
- A complete graph made of five nodes
- A bipartite graph made of a two-node group and a three-node group in which all possible inter-group edges are present
- A bipartite graph made of two three-node groups in which all possible intergroup edges are present

```python
import networkx as nx
from networkx.algorithms import bipartite
from pylab import *

print('A complete graph made of four nodes')
g1 = nx.complete_graph(4)
nx.draw(g1)
show()

print('A complete graph made of five nodes')
g2 = nx.complete_graph(5)
nx.draw(g2)
show()

print("""A bipartite graph made of a two-node group and a three-node group
in which all possible inter-group edges are present""")
g3 = nx.Graph()
g3.add_nodes_from([1, 2], bipartite=0)
g3.add_nodes_from(['a', 'b', 'c'], bipartite=1)
g3.add_edges_from([(1,'a'),(1,'b'),(1,'c'),(2,'a'),(2,'b'),(2,'c')])
nx.draw(g3, node_color = ['yellow' if i in [1,2] else 'green' for i in g3.nodes()])
show()
print('is_bipartite: ', bipartite.is_bipartite(g3))

print("""\nA bipartite graph made of two three-node groups
in which all possible intergroup edges are present""")
g4 = nx.Graph()
g4.add_nodes_from([1, 2, 3], bipartite=0)
g4.add_nodes_from(['a', 'b', 'c'], bipartite=1)
g4.add_edges_from([(1,'a'),(1,'b'),(1,'c'),
                   (2,'a'),(2,'b'),(2,'c'),
                   (3,'a'),(3,'b'),(3,'c')])
nx.draw(g4, node_color = ['yellow' if i in [1,2,3] else 'green' for i in g4.nodes()])
show()
print('is_bipartite: ', bipartite.is_bipartite(g4))
```

## Sayama Exercise 15.4
Every connected tree graph made of $n$ nodes has exactly $n−1$ edges. Explain why

The minimum number of edges required to make a connected graph of $n$ vertices is $(n-1)$ edges. If we picture having two nodes than the number of edges required to connect them is 1. For three nodes, the number of edges required to connect them is 2. Consequently, if the gap between the number of edges and nodes is more than one, then the tree would be disconnected.
